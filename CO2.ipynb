```python
import pandas as pd
import numpy as np
```

```python
file_to_load = open("data/CO2_emission_by_countries.csv", encoding='utf8', errors='ignore')
file_to_load
```

```python
carbon_df = pd.read_csv(file_to_load)
carbon_df
```

```python
carbon_df = carbon_df.drop(columns= {"Calling Code", "Area", "Density(km2)"})
carbon_df
```

```python
file_to_load= open("data/countryContinent.csv", encoding='utf8', errors='ignore')
country_cont_df = pd.read_csv(file_to_load)
country_cont_df 
```

```python
country_cont_df = country_cont_df.drop(columns=['country','code_3', 'country_code', 'iso_3166_2', 'region_code','sub_region_code'])
country_cont_df
```

```python
country_cont_df.rename(columns = {"code_2":"Code"}, inplace=True)
```

```python
country_cont_df
```

```python
CO2_emissions_DF = pd.merge(carbon_df, country_cont_df, left_on='Code', right_on='Code')
CO2_emissions_DF
```

```python
CO2_emissions_DF=CO2_emissions_DF.dropna()
CO2_emissions_DF
```

```python
CO2_emissions_DF.columns
```

```python
CO2_emissions_DF
```

```python
samples = CO2_emissions_DF[['Country', 'Year', 'CO2 emission (Tons)']]
# samples = samples.T
#samples = samples.set_index('Country')
#samples = samples.T 
samples
```

```python
samples_cleaned = samples.loc[samples['CO2 emission (Tons)']!= 0]
samples_cleaned = samples_cleaned.set_index('Country')
samples_cleaned = samples_cleaned.reset_index()
samples_cleaned
```

```python
samples_cleaned['Year'] = samples_cleaned['Year'].values.astype(str)
```

```python
samples_cleaned = samples_cleaned.pivot(index='Country', columns='Year', values='CO2 emission (Tons)')
samples_cleaned
```

```python
samples_cleaned = samples_cleaned.loc[:, '1949':'2020']
```

```python
samples_cleaned = samples_cleaned.reset_index()
samples_cleaned = samples_cleaned.dropna()
samples_cleaned
```

```python
names = samples_cleaned['Country']
names 
```

```python
names_df = pd.DataFrame(names)
```

```python
names_df
```

```python
names_df.T.to_json(r"CO2_names.json", orient='values')
```

```python
metadata = CO2_emissions_DF[CO2_emissions_DF['Year']>=1949]
metadata 
```

```python
metadata  = pd.merge(names_df, metadata, left_on='Country', right_on='Country')
metadata 
```

```python
metadata = metadata.drop_duplicates().reset_index()
metadata  = metadata [['Country', 'Code','Population(2022)','% of World', 'continent', 'sub_region']]
metadata
```

```python
metadata.to_json("CO2_metadata.json", orient='records')
```
