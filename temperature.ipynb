```python
#country_temp_df_dict = country_temp_df.T.to_dict()

country_temp_new = []
for x,vals in country_temp_df_dict.items():
    for y, z in vals.items():
        country_temp_new.append([x,y,z])
        
country_temp_df_new = pd.DataFrame(country_temp_new, columns=[ "Country", "Year", "Temp"])
country_temp_df_new 
```

```python
from matplotlib.pyplot import figure
plt.figure(figsize = (20,5))
labels = country_temp_df_new['Year'] 
countries = country_temp_df_new['Country']
temps = country_temp_df_new['Temp']

plt.scatter(labels, temps)
plt.show()
```
