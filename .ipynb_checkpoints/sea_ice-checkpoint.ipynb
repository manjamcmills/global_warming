```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sea_ice_NH_extent = pd.read_csv("data/sea_ice_NH_extent.csv")
sea_ice_NH_extent.drop(["Unnamed: 13"], axis=1, inplace=True)
sea_ice_NH_extent.rename(columns = {"Unnamed: 0":"year"}, inplace=True)
sea_ice_NH_extent.rename(columns = {"Annual":"Annual Sea Ice NH Extent"}, inplace=True)
sea_ice_NH_extent.set_index('year', inplace=True)
sea_ice_NH_extent.head()
```

```python
sea_ice_NH_extent_dict = sea_ice_NH_extent.drop(['Annual Sea Ice NH Extent'], axis=1).T.to_dict()

sea_ice_NH_extent_mo  = []
for x, vals in sea_ice_NH_extent_dict.items():
    for y, z in vals.items():
        sea_ice_NH_extent_mo.append([x,y,z])
        
sea_ice_NH_extent_DF = pd.DataFrame(sea_ice_NH_extent_mo, columns=["Year", "Month", "Sea Ice NH Extent"])
sea_ice_NH_extent_DF
```

```python
sea_ice_NH_extent_annual=pd.DataFrame(sea_ice_NH_extent, columns=['Annual Sea Ice NH Extent'])
sea_ice_NH_extent_annual.rename(columns = {"Annual Sea Ice NH Extent":"Sea Ice NH Extent"}, inplace=True)
sea_ice_NH_extent_annual.head() 
```

```python
sea_ice_NH_area = pd.read_csv("data/sea_ice_NH_area.csv")
sea_ice_NH_area.drop(["Unnamed: 13"], axis=1, inplace=True)
sea_ice_NH_area.rename(columns = {"Unnamed: 0":"year"}, inplace=True)
sea_ice_NH_area.rename(columns = {"Annual":"Annual Sea Ice NH Area"}, inplace=True)
sea_ice_NH_area.set_index('year', inplace=True)
sea_ice_NH_area.head()
```

```python
sea_ice_NH_area_dict = sea_ice_NH_area.drop(['Annual Sea Ice NH Area'], axis=1).T.to_dict()

sea_ice_NH_area_dict.items()
```

```python
sea_ice_NH_area_dict = sea_ice_NH_area.drop(['Annual Sea Ice NH Area'], axis=1).T.to_dict()

sea_ice_NH_area_mo = []
for x,vals in sea_ice_NH_area_dict.items():
    for y, z in vals.items():
        sea_ice_NH_area_mo.append([x,y,z])
        
sea_ice_NH_area_DF = pd.DataFrame(sea_ice_NH_area_mo, columns=["Year", "Month", "Sea Ice NH Area"])
sea_ice_NH_area_DF
```

```python
sea_ice_NH_area_annual=pd.DataFrame(sea_ice_NH_area, columns=['Annual Sea Ice NH Area'])
sea_ice_NH_area_annual.rename(columns = {"Annual Sea Ice NH Area":"Sea Ice NH Area"}, inplace=True)
sea_ice_NH_area_annual.head() 
```

```python
sea_ice_SH_extent = pd.read_csv("data/sea_ice_SH_extent.csv")
sea_ice_SH_extent.drop(["Unnamed: 13"], axis=1, inplace=True)
sea_ice_SH_extent.rename(columns = {"Unnamed: 0":"year"}, inplace=True)
sea_ice_SH_extent.rename(columns = {"Annual":"Annual Sea Ice SH Extent"}, inplace=True)
sea_ice_SH_extent.set_index('year', inplace=True)
sea_ice_SH_extent.head()
```

```python
sea_ice_SH_extent_dict = sea_ice_SH_extent.drop(['Annual Sea Ice SH Extent'], axis=1).T.to_dict()

sea_ice_SH_extent_mo  = []
for x, vals in sea_ice_SH_extent_dict.items():
    for y, z in vals.items():
        sea_ice_SH_extent_mo.append([x,y,z])
        
sea_ice_SH_extent_DF = pd.DataFrame(sea_ice_SH_extent_mo, columns=["Year", "Month", "Sea Ice SH Extent"])
sea_ice_SH_extent_DF
```

```python
sea_ice_SH_extent_annual=pd.DataFrame(sea_ice_SH_extent, columns=['Annual Sea Ice SH Extent'])
sea_ice_SH_extent_annual.rename(columns = {"Annual Sea Ice SH Extent":"Sea Ice SH Extent"}, inplace=True)
sea_ice_SH_extent_annual.head() 
```

```python
sea_ice_SH_area = pd.read_csv("data/sea_ice_SH_area.csv")
sea_ice_SH_area.drop(["Unnamed: 13"], axis=1, inplace=True)
sea_ice_SH_area.rename(columns = {"Unnamed: 0":"year"}, inplace=True)
sea_ice_SH_area.rename(columns = {"Annual":"Annual Sea Ice SH Area"}, inplace=True)
sea_ice_SH_area.set_index('year', inplace=True)
sea_ice_SH_area.head()
```

```python
sea_ice_SH_area_dict = sea_ice_SH_area.drop(['Annual Sea Ice SH Area'], axis=1).T.to_dict()

sea_ice_SH_area_mo = []
for x,vals in sea_ice_SH_area_dict.items():
    for y, z in vals.items():
        sea_ice_SH_area_mo.append([x,y,z])
        
sea_ice_SH_area_DF = pd.DataFrame(sea_ice_SH_area_mo, columns=["Year", "Month", "Sea Ice SH Area"])
sea_ice_SH_area_DF
```

```python
sea_ice_SH_area_annual=pd.DataFrame(sea_ice_SH_area, columns=['Annual Sea Ice SH Area'])
sea_ice_SH_area_annual.rename(columns = {"Annual Sea Ice SH Area":"Sea Ice SH Area"}, inplace=True)
sea_ice_SH_area_annual.head() 
```

```python
sea_ice_annual_list = [sea_ice_NH_extent_annual,sea_ice_NH_area_annual,sea_ice_SH_extent_annual,sea_ice_SH_area_annual]
```

```python
sea_ice_annual_merged = sea_ice_NH_extent_annual.merge(sea_ice_NH_area_annual, on ='year').merge(sea_ice_SH_extent_annual, on='year').merge(sea_ice_SH_area_annual, on='year')
```

```python
sea_ice_annual_merged = sea_ice_annual_merged.rename(columns={'Sea Ice NH Extent':'NH Extent', 'Sea Ice NH Area':'NH Area','Sea Ice SH Extent':'SH Extent', 'Sea Ice SH Area':'SH Area' })
sea_ice_annual_merged.head()
```

```python
sea_ice_annual_merged.columns
```

```python
sea_ice = sea_ice_annual_merged
sea_ice 
```

```python
sea_ice = sea_ice.rename(columns = {'NH Extent':'NH_Extent', 'NH Area':'NH_Area', 'SH Extent':'SH_Extent', 'SH Area':'SH_Area'})
sea_ice
```

```python
sea_ice.columns
```

```python
sea_ice_global = sea_ice.copy()
sea_ice_global['Global Sea Ice Area'] = sea_ice_global.loc['1978':'2022', ['NH_Area',  'SH_Area']].sum(axis=1)
sea_ice_global['Global Sea Ice Extent'] = sea_ice_global.loc['1978':'2022', ['NH_Extent',  'SH_Extent']].sum(axis=1)
sea_ice_global.head()
```

```python
sea_ice_global=sea_ice_global['Global Sea Ice'] 
sea_ice_global
```

```python
sea_ice_global.to_json("global_sea_ice.json")
```

```python
from sklearn.linear_model import LinearRegression
sea_ice_SH_area_annual =sea_ice_SH_area_annual.reset_index()
X = sea_ice_SH_area_annual.year.values.reshape(-1, 1)
y = sea_ice_SH_area_annual['Sea Ice SH Area']
model_SH_area = LinearRegression()
model_SH_area.fit(X, y)
y_pred_SH_area  = model_SH_area.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred_SH_area , color='red')
plt.show()
```

```python
print(model_SH_area.coef_)
print(model_SH_area.intercept_)
```
