```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
sealevels_mean = sealevels_mean.drop(columns={'ObjectId', 'Country', 'ISO2', 'ISO3', 'Indicator', 'Unit', 'Source',
       'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor',})
sealevels_mean 
```

```python
import re
sealevels_mean['Date'] = sealevels_mean['Date'].str.replace('D','')
```

```python
sealevels_mean['Date'] = pd.to_datetime(sealevels_mean['Date'])
sealevels_mean
```

```python
sealevels_mean['Measure'].value_counts()
```

```python
world_sea_level = sealevels_mean.copy()
world_sea_level = world_sea_level.loc[(sealevels_mean['Measure']=='World')]
world_sea_level = world_sea_level.drop(columns='Measure').rename(columns = {'Value':'World'})
world_sea_level = world_sea_level.reset_index()
world_sea_level
```

```python
world_sea_level = world_sea_level.groupby(["Date"]).mean()
world_sea_level
```

```python
atl_ocean = sealevels_mean.loc[(sealevels_mean['Measure']=='Atlantic Ocean')]
atl_ocean = atl_ocean.drop(columns='Measure').rename(columns = {'Value':'Atlantic Ocean'})
atl_ocean = atl_ocean.reset_index()
atl_ocean
```

```python
atl_ocean = atl_ocean.drop(columns={'index'})
atl_ocean
```

```python
atl_ocean = atl_ocean.groupby(["Date"]).mean()
atl_ocean
```

```python
pac_ocean = sealevels_mean.loc[(sealevels_mean['Measure']=='Pacific Ocean')]
pac_ocean = pac_ocean.drop(columns='Measure').rename(columns = {'Value':'Pacific Ocean'})
pac_ocean = pac_ocean.reset_index()
pac_ocean
```

```python
pac_ocean = pac_ocean.drop(columns={'index'})
pac_ocean
```

```python
pac_ocean= pac_ocean.groupby(["Date"]).mean()
pac_ocean
```

```python
ind_ocean = sealevels_mean.loc[(sealevels_mean['Measure']=='Indian Ocean')]
ind_ocean = ind_ocean.drop(columns='Measure').rename(columns = {'Value':'Indian Ocean'})
ind_ocean = ind_ocean.reset_index()
ind_ocean= ind_ocean.drop(columns={'index'})
ind_ocean
```

```python
ind_ocean= ind_ocean.groupby(["Date"]).mean()
ind_ocean
```

```python
nino = sealevels_mean.loc[(sealevels_mean['Measure']=='Nino')]
nino = nino.drop(columns='Measure').rename(columns = {'Value':'Nino'})
nino  =nino.reset_index()
nino= nino.drop(columns={'index'})
nino 
```

```python
nino = nino .groupby(["Date"]).mean()
nino 
```

```python
tropics = sealevels_mean.loc[(sealevels_mean['Measure']=='Tropics')]
tropics= tropics.drop(columns='Measure').rename(columns = {'Value':'Tropics'})
tropics = tropics.reset_index()
tropics= tropics.drop(columns={'index'})
tropics
```

```python
tropics=tropics.groupby(["Date"]).mean()
tropics
```
