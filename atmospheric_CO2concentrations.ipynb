```python
#importing packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
atmosphere_warming = pd.read_csv('data/atmospheric_CO2concentrations.csv')
```

```python
atmosphere_warming['Parts Per Million'] = np.where(atmosphere_warming['Unit']!= 'Percent', True, False)
atmosphere_warming
```

#Sorting the 'Unit' column by Parts Per Million | Percent to extract the correct temp values from the 'Value' column

```python
PPM_CO2 = atmosphere_warming[atmosphere_warming['Parts Per Million'] == True]
Percent_CO2 = atmosphere_warming[atmosphere_warming['Parts Per Million'] == False]


PPM_CO2
```

```python
Percent_CO2
```

```python
Percent_CO2 = Percent_CO2.rename(columns={'Value': 'CO2_percent'})
```

```python
Percent_CO2.style.format({'CO2_percent': "{:.2%}"})
```

```python
Percent_CO2.style.format({'CO2_percent': "{:.2%}"})
Percent_CO2.set_index('Year', inplace=True)
Percent_CO2.groupby('Unit')['CO2_percent'].plot(legend=True)
plt.savefig('Percent_CO2.png')
plt.show()
Percent_CO2
```

```python
PPM_CO2
```

```python
PPM_CO2.set_index('Year', inplace=True)
PPM_CO2.groupby('Unit')['Value'].plot(legend=True)
plt.savefig('PPM_CO2.png')
plt.show()
PPM_CO2
```
